```python
import time
import re
from bs4 import BeautifulSoup
click_text = "#pageObjNavgation > div > span > a:nth-child({})"
p = re.compile("melon.link.goSongDetail\('([0-9]+)'\)")
total_id  = []
for page_number in range(1,100):
    try:
        driver.find_element(By.CSS_SELECTOR,
                click_text.format(page_number)).click()
        time.sleep(3)
    except:
        if page_number == 1:
            pass
        else:
            break
    bs = BeautifulSoup(driver.page_source)
    imsi  = bs.find_all("form", id="frm_defaultList")[0]
    songid = [p.findall(str(x))[0] for x in imsi.find_all("a", class_="btn btn_icon_detail")]
    total_id.extend(songid)
```

```python
import requests
def get_lyrics(id_):
    global r
    head = {"User-Agent":
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"}
    url = "https://www.melon.com/song/detail.htm?songId={}"
    r = requests.get(url.format(id_),headers=head)
    song_tex = BeautifulSoup(r.text)
    lyrics = BeautifulSoup(str(song_tex.find("div", class_="lyric")).replace("<br/>", "\n")).text.strip()
    title, open_dt, type_, _ =[x.text for x in song_tex.find("dl", class_='list').findAll("dd")]
    song_title = song_tex.find("div", class_="song_name").text.split()[-1]
    singer = song_tex.find("div", class_="artist").find("a").find("span").text
    return singer, song_title, lyrics, title, open_dt, type_
```
