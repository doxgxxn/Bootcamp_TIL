```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
name=[]
category=[]
addr=[]
number=[]
lat=[]
lon=[]
eng_addr=[]
```

```python
r_list = [name,category,addr,number,lat,lon,eng_addr]
```

```python
col_name = ["name","category","addr","number","lat","lon"]
```

```python
from bs4 import BeautifulSoup
url = "https://guide.michelin.com/kr/ko/seoul-capital-area/kr-seoul/restaurants/page/{}"

for i in range(1, 11):
    r = requests.get(url.format(i))
    soup = BeautifulSoup(r.text, "html.parser")
    section = soup.find_all('div', class_="col-md-6 col-lg-4 col-xl-3")

    for sec in section:
        name.append(' '.join(sec.a['aria-label'].split()[1:]))
        category.append(sec.text.strip().split()[-1])
        lat.append(sec.div['data-lat'])
        lon.append(sec.div['data-lng'])
        eng_addr.append(sec.a['href'].split('/')[-1])
```

```python
import re
import numpy as np
import tqdm
```

```python
from bs4 import BeautifulSoup
r_url = "https://guide.michelin.com/kr/ko/seoul-capital-area/kr-seoul/restaurant/{}"

for restaurant in tqdm.tqdm(eng_addr):
    r = requests.get(r_url.format(restaurant))
    soup = BeautifulSoup(r.text, 'html.parser')

    addr.append(soup.find('li', class_="restaurant-details__heading--address").text.split(',')[:-2][0])
    try:
        pn = "0" + soup.find('a', href=re.compile(r'^tel'))['href'].split()[1]
        number.append(pn)
    except:
        number.append(np.nan)

    
```
